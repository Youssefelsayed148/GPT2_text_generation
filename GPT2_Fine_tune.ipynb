```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
```

```python
checkpoint= 'gpt2'
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForCausalLM.from_pretrained(checkpoint).to(device)
```

```python
model.eval()
```

```python
prompt= "She opened the ancient book and discovered"
input_ids = tokenizer(prompt, return_tensors="pt")["input_ids"].to(device)
```

```python
outputs= model.generate(input_ids,
                        max_length=100,
                        do_sample=True,
                        top_k=50,
                        top_p=0.95,
                        temperature=0.7,
                        num_return_sequences=2)
```

```python
for i, output in enumerate(outputs):
    print(f"Generated sequence {i+1}:")
    print(tokenizer.decode(output, skip_special_tokens=True))
```

```python
prompt= "Unpopular opinion:"
input_ids = tokenizer(prompt, return_tensors="pt")["input_ids"].to(device)
outputs_2= model.generate(input_ids,
                        max_length=80,
                        do_sample=True,
                        top_k=50,
                        top_p=0.95,
                        temperature=0.7,
                        num_return_sequences=2,
                        no_repeat_ngram_size=2)


for i, output in enumerate(outputs_2):
    print(f"Generated sequence {i+1}:")
    print(tokenizer.decode(output, skip_special_tokens=True))
```

```python
prompt= ["Machine learning will revolutionize","The human brain is capable of","Here is a Python function that adds two numbers:","The best advice I ever got was"]
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = 'left'
input_ids = tokenizer(prompt, padding=True,return_tensors="pt")["input_ids"].to(device)
outputs_3= model.generate(input_ids,
                        max_length=80,
                        do_sample=True,
                        top_k=50,
                        top_p=0.95,
                        temperature=0.7,
                        num_return_sequences=1,
                        no_repeat_ngram_size=2)

results=[]
for i, output in enumerate(outputs_3):
    print(f"Generated sequence {i+1}:")
    decoded=tokenizer.decode(output, skip_special_tokens=True)
    print(decoded)
    results.append({"prompt": prompt[i],
                     "generated": decoded
    })
```
